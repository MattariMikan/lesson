# if文（条件分岐）
条件に応じて異なる処理を実行

````python
#xは整数
x=6
if x>=5:
  print("5以上")
elif x>=1:
  print("1以上")
else:
  print("0")
````

・リスト<br>

# for文（反復処理）
ある処理を繰り返し実行

````python
x= [1, 2, 3, 4, 5]
for i in x:
  print(i)
````

# 関数
あるまとまった処理を実行するための仕組み<br>



````python
#足し算の結果を返す関数　returnは戻り値（返される値）
def wa(x, y):
  return x+y

print(wa(1, 3))
````
